-- 배열의 모양 변경

```python
# 12개의 요소를 가지는 1차원 행렬 생성
a = np.arange(1,13,1) 
print(a)

# reshape()
# : 행열의 크기를 괄호안의 값에 맞게 바꾸어준 행렬을 리턴해줌
# : 단, 원래 요소의 개수에 맞춰서 행과 열의 개수를 정해줘야함.
a1= a.reshape(3,4)
print(a1)

# resize()
# : 리턴값은 없고, 바로 a의 모양을 바로 바꾸어 저장해줌
a.resize(6,2)
print(a)

# flatten()
# : 1차원 배열로 변경된 배열을 리턴(즉 a를 바꾸고 싶으면 다시 저장해야됨.)
a = a.flatten()
print(a)

# transpose()
# : 두가지가 같은 결과가 나옴
print(a1.T)
print(a1.transpose()) # 참고로 괄호안에 축의 순서를 지정할 수 있음
```

```python
# 모든 초콜릿의 평점 계산
ratings_mean=data[:,3].mean()
print("모든 초콜릿의 평점 : ", ratings_mean)

high_level = data[data[:,3] >=4]
print(high_level.shape)
```

```python
# 평점이 4점이상인 초콜릿제품의 수 계산
print(high_level.shape[0])

high_id = high_level[:,0].astype(np.int64)
print(high_id)
print(f'우수 초콜릿의 수 {len(high_id)}, {high_id.size}')
```

```python
# 카카오 함유량 의 빈도 계산
high_kakao = high_level[:,2]
print(high_kakao)

# 중복삭제
unique_values, value_count = np.unique(high_kakao, return_counts=True)
#print(unique_values, value_count)

t=list(zip(unique_values, value_count))
print(t)

for k, v in t:
    print(k, v)
```

```python
# 카카오 함유량 빈도수가 가장 높은 함유량
aa = np.array([1,5,22,6])
# np.argmax() : 배열에서 최대값의 인덱스(위치)를 나타냄
print(np.argmax(aa))

# 실제 적용
max_index = np.argmax(value_count)
print(max_index)   # 8번째 인덱스의 값이 현재 45로 최대임
max_kakao = unique_values[max_index]
print(f'우수 초콜릿 {len(high_id)} 가지 중 카카오 함유량 {max_kakao}이(가)', end=" ") 
print(f'빈도 {value_count[max_index]}번으로 가장 많습니다. ')
```

```python
# 사용자별 영화평점의 평균
user_ids= np.unique(data[:,0])
#print(len(user_ids))

mean_values=[]
for user_id in user_ids:
    # data[조건] 조건에 맞는 값을 모두 불러올 수 있음
    data_user = data[data[:,0] == user_id]
    value = data_user[:,2].mean()
    mean_values.append([user_id, value])

mean_values    
```

```python
print(len(mean_values))
mean_array=np.array(mean_values)
print(mean_array.shape)
np.savetxt("data/result.csv", mean_array, fmt="%.1f", delimiter=",")
```

```python
arr = np.array(data).T; arr
df2 = pd.DataFrame(arr, columns=["names", "years", "points"], 
                          index=['one', 'two', 'three', 'four', 'five'])
print(df2)
df2
# 인덱스를 객체리스트로 저장
print(df2.index)
# 칼럼이름을 객체리스트로 저장
print(df2.columns)
# 2차원 배열값으로 보여줌
print(df2.values)
```

```python
#데이터 프레임의 칼럼을 하나 뽑아오게 되면 시리즈로 저장됨
names = df2['names']
years = df2['years']
points = df2['points']
print(names)
print(years)
print(points)
```
