# bool 자료형

```python
# []안에 아무것도 입력안하면 false
if [] :
    print('aaa')
else :
    print('bbb')
```

```python
#''안에 아무것도 안적으면 flase
if '' :
    print("참")
else : 
    print("거짓")
```

```python
# 1은 참
if 1 :
    print("참")
else : 
    print("거짓")

# 0은 거짓
if 0 :
    print("참")
else : 
    print("거짓")
```
