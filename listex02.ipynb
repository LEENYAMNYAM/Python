<추가, 삭제, 변경 등등등.... 기본입출력 관련 함수>

```python
# .append(값) : 요소 한개추가
a.append(6)
print(a)
a.append([7,8])         
print(a)

# .extend([값, 값, 값, ... ]) : 요소 여러개 추가
a.extend([9,10])
print(a)

# .insert(인덱스, 값) : 요소를 원하는 위치에 추가
a.insert(1,100)
print(a)

# del 리스트명[인덱스] : 인덱스에 해당하는 요소 삭제
del a[1]
print(a)

# .remove(값) : 값을 가지는 요소 삭제 
a.remove(2)
print(a)

# .reverse() : 요소의 순서를 역순으로 바꿈
print(a[::-1])
a.reverse()
print(a)
```

```python
# .index() : 인덱스 값을 불러와줌
a.index(3)
    # 없는 인덱스를 입력하면 에러가 생김
#a.index(11)
```

```python
# .sort() : 값의 크기순(기본은 오름차순)으로 정렬함
a=[4,2,5,1,3]
a.sort()
print(a)
    # 문자도 마찬가지 알파벳순 정렬
s1=["가나다", "정현민", "조유빈", ""]
s1.sort()
print(s1)
```

```python
# .pop() : 끝값부터 하나씩 꺼내옴

b=[10,20,30,40,50]
print(b)
print(b.pop()) #마지막값 50이 깨내어짐
print(b)
```

```python
# .count() : 글자의 개수 파악
b.append(10)
print(b)
print(b.count(10))  # 10이라는 값을 2개 가지고 있으므로 2가 출력됨.
```

TMI

```python
import re

# Original string
text = "Tesla-sales are.in an-unprecedented.slide"

# Split on multiple delimiters
words = re.split(r"[.\-\s]", text)

# Remove empty strings caused by consecutive delimiters
words = [word for word in words if word]

print(words)
# Output: ['Tesla', 'sales', 'are', 'in', 'an', 'unprecedented', 'slide']
```
